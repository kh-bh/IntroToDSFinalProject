```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
YTSHighdf = pd.read_csv("../YTS Data\YTSCessationData.csv")
YTSMiddledf = pd.read_csv("../YTS Data\YTSMiddleData.csv")
YTSSmokelessTobaccodf = pd.read_csv("../YTS Data\YTSSmokelessTobaccoData.csv")
YTSCigaretteUsedf = pd.read_csv("../YTS Data\YTSCigaretteUseData.csv")
YTSCessationdf = pd.read_csv("../YTS Data\YTSCessationData.csv")
```

```python
#creating graph for Cessation Data Bhoomika
years = YTSCigaretteUsedf['Year'].unique()
datavalue = YTSCessationdf.groupby(['MeasureDesc','Year']).agg({'DataValue': ['mean']})

stderror = YTSCessationdf.groupby(['MeasureDesc','Year']).agg({'DataValueStdErr': ['mean']})
meansstd = stderror['DataValueStdErr']
#print (meansstd)
plt.figure(figsize=(10, 6))
plt.plot(years,datavalue[0:19],label='label1', c = "seagreen", marker = "o")
plt.plot(years,datavalue[19:],label='label1', c = "orange", marker = "o")
plt.xticks(years)
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True)

plt.legend(["Percent of Current Smokers Who Want to Quit" , "Quit Attempt in Past Year Among Current Cigarette Smokers"])

plt.title('Cigarette Cessastion by Year')
plt.xlabel('Year')
plt.ylabel('Mean Percent of Cessation')

#plt.errorbar(years, datavalue[0:19], yerr=meansstd[0:19], fmt="o")
```

```python
import plotly.express as px
YTSCessationdf["std"] = YTSCessationdf["DataValueStdErr"]/100
fig = px.scatter(YTSCessationdf, x="Year", y="DataValue", color="MeasureDesc",
                 error_y="std")
fig.show()
```
