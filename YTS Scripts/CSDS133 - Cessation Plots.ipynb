```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import csv
```

```python
YTSHighdf = pd.read_csv("../YTS Data\YTSCessationData.csv")
YTSMiddledf = pd.read_csv("../YTS Data\YTSMiddleData.csv")
YTSSmokelessTobaccodf = pd.read_csv("../YTS Data\YTSSmokelessTobaccoData.csv")
YTSCigaretteUsedf = pd.read_csv("../YTS Data\YTSCigaretteUseData.csv")
YTSCessationdf = pd.read_csv("../YTS Data\YTSCessationData.csv")
```

```python
#creating graph for Cigarette Use by year Monish
years = []
cigarette_use = []

years = YTSCigaretteUsedf['Year'].unique()
cigarette_use = YTSCigaretteUsedf.groupby('Year')['DataValue'].mean()

plt.figure(figsize=(10, 6))
plt.plot(years, cigarette_use, marker='o')
plt.xlabel('Year')
plt.ylabel('Cigarette Use')
plt.title('Cigarette Use by Year')
plt.grid(True)
plt.xticks(years)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
